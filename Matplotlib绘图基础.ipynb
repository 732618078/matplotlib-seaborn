# 堆积图

```python
comp = [1, 2, 3, 4]
y1 = [24, 27, 35, 43]
y2 = [30, 35, 17, 12]
y3 = [27, 13, 15, 8]
y4 = [19, 25, 33, 37]
```

```python
plt.rcParams['font.sans-serif'] = ['SimHei'] #用来显示中文
plt.rcParams['axes.unicode_minus'] = False  #用来显示符号
```

```python
labels = ['apple', 'micro', 'yahoo', 'google']
plt.stackplot(comp, y1, y2, y3, y4, labels=labels)
plt.legend()
plt.xticks([1, 2, 3, 4], ['2011年', '2012年', '2013年', '2014年']); #修改刻度名称
```

# 面向对象绘图

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
y = np.sin(x)
y1 = np.cos(x)
```

```python
plt.plot(x, y, label='sin')
plt.plot(x, y1, label='cos')
plt.xlim(-4, 4)
plt.ylim(-1, 1) #设置坐标轴的范围
plt.xticks([-np.pi, 0, np.pi], [r'$-\pi$', 0, r'$+\pi$'])
plt.legend(loc=0);
```

```python
plt.plot(x, y, label='sin')
plt.plot(x, y1, label='cos')
plt.axis([-3.2, 3.2, -1, 1])
plt.xticks([-np.pi, 0, np.pi], [r'$-\pi$', 0, r'$+\pi$']) #更改坐标轴刻度
plt.legend(loc=0);
```

# 边框处理和水平线条

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
y = np.sin(x)
```

```python
fig ,ax = plt.subplots()
ax.plot(x, y)
ax.spines['top'].set_visible(False) #设置特定边框不可见
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('r') #设置特定边框颜色
ax.spines['left'].set_linewidth(5) #设置特定边框线宽
ax.spines['bottom'].set_position(('data', 0)) #设置特定边框位置
ax.spines['left'].set_position(('data', 0))
ax.axhline(1, color='r', linewidth=1) #画水平线
ax.axvline(1.5, color='y', linewidth=1); #画垂直线
```

# 共享轴

```python
x = [1, 2, 3, 4]
x1 = [3, 4, 5, 6]
y = [4, 5, 8, 9]
y1 = [3, 6, 2, 5]
```

```python
fig, ax = plt.subplots(2, 1, sharex=True, sharey=True)
ax[0].bar(x, y)
ax[0].bar(x, y1, bottom=y)
ax[1].plot(x, y);
```

```python
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(2, 1, 2, sharex=ax1, sharey=ax1)
ax1.plot(x, y)
ax2.plot(x1, y1)
plt.setp(ax1.get_xticklabels(), visible=False); #不显示刻度值
```

```python
fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(2, 1, 2, sharex=ax1, sharey=ax2)
ax1.plot(x, y)
ax2.plot(x1, y1)
for lab in ax1.get_xticklabels():
    lab.set_visible(False);
```
